```python
from Bio import SeqIO
import numpy as np
import random
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import subprocess
from Bio import SearchIO
import pandas as pd
import re
import math
import argparse

parser = argparse.ArgumentParser(description='Peptide data augmentation')
parser.add_argument('--fasta',metavar='fasta', type=str, help='Fasta file')
parser.add_argument('--output',metavar='output', type=str, help='Output file')
args = parser.parse_args()

def seq_prob_list(alp):
    seq_list = []
    aa_group = [['R','H','K'],\
            ['D','E'],\
            ['S','T','N','Q'],\
            ['C','G','P'],\
            ['A','V','I','L','M','F','Y','W']]
    if random.uniform(0,1)<0.6 and alp!='X':
        for l in aa_group:
            if alp in l:
                l.remove(alp)
                return random.sample(l,1)
    else:
        for l in aa_group:
            if not alp in l:
                seq_list+=l

        return random.sample(seq_list,1)
            


def rand_seq_gen(sequence):

    substitute_count = 0
    miss_count=-1
    seq = []
    substitute_count = 0
    miss_count+=1
    for i, alp in enumerate(sequence):
        ## Random AA addition
        if random.uniform(0,1)<0.1:
            seq.extend(seq_prob_list('X'))
            substitute_count+=1

        ## Random AA substitution
        if i==0 and alp=='M':
            seq.extend(alp)
        elif random.uniform(0,1)<0.5:
            seq.extend(seq_prob_list(alp))
            substitute_count+=1
        else:
            seq.extend(alp)
    
    print(seq)
    return ''.join(seq)
        
print('Read original sequence data file--------------')
orig_seq_data = SeqIO.parse(args.fasta,'fasta')
data = []
id_info = []
for l in orig_seq_data:
    data.append(l)
```
